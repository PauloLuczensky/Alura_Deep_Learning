## **Equação da reta**

A equação da reta no formato *ax + by + c* será reeescrita para ***w1x1 + w2x2+b***, Ou seja, w1, w2 e b são parâmetros que definem o nosso modelo linear.




```python
#Plotando uma reta -1x1 + 4x2 + 0.4
import numpy as np
import matplotlib.pyplot as plt


a = -1
b = 4
c= 0.4

# ax + by + c
def plotline(a, b, c):
  x = np.linspace(-2, 4, 50)
  y = (-a * x - c)/b
  print(x)

  plt.axvline(0, -1, 1, color = 'k', linewidth = 1)
  plt.axhline(0, -2, 4, color = 'k', linewidth = 1)
  plt.plot(x, y)
  plt.grid(True)

plotline(a, b, c)
```

##**Solucionando a equação para diferentes pontos**



```python
p1 = (2, 0.4)
p2 = (1, 0.6)
p3 = (3, -0.4)

ret1 = a*p1[0] + b*p1[1] + c

plotline(a, b, c)
plt.plot(p1[0], p1[1], color='b', marker='o')
plt.plot(p2[0], p2[1], color='r', marker='o')
plt.plot(p3[0], p3[1], color='g', marker='o')
```

```python
#Eq da reta: ax + by + c = 0
#Em cima da reta, o resultado vai ser 0
ret1 = a*p1[0] + b*p1[1] + c
print("%.2f" % ret1)

#Qualquer ponto acima da reta, o valor é +
ret2 = a*p2[0] + b*p2[1] + c
print("%.2f" % ret2)

#Qualquer ponto abaixo da reta, o valor é -
ret3 = a*p3[0] + b*p3[1] + c
print("%.2f" % ret3)
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(46)

#x -> armazena todas as dimensões do dado, neste casos 2 features(x e y)
#y -> armazena os labels, se pretence a classe 0 ou 1
X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)

print(X.shape, Y.shape)

#plotando as duas classes, os dados de x que pertencem a classe 0 e 1
plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolor='k')

#Plotando a classe que pertence o ponto na posição 10
p = X[10]
print(X[10])
print(Y[10])
plt.plot(p[0], p[1], marker='^', markersize=20)
```

```python
def plotmodel(w1, w2, b):

  plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolor='k')

  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  x = np.linspace(-2, 4, 50)
  y = (-w1 * x -b)/w2
  print(x)

  plt.axvline(0, -1, 1, color = 'k', linewidth = 1)
  plt.axhline(0, -2, 4, color = 'k', linewidth = 1)
  plt.plot(x, y)
  plt.grid(True)

  plt.xlim(xmin, xmax)
  plt.ylim(ymin, ymax)

w1 = 5
w2 = 1
b = -0.4

plotmodel(w1,w2,b)

#Sabendo o lado + e - da minha reta
p = (-1,1)
print(w1*p[0] + w2*p[1] + b)
#valores - em relação a reta é azul
```

```python
def classify(ponto, w1, w2, b):
  ret = w1 * ponto[0] + w2 * ponto[1] + b

  if ret >= 0:
    return 1, 'yellow'
  else:
    return 0, 'blue'


p = (2, -1)
classe, color = classify(p, w1, w2, b)
print(classe, color)

plotmodel(w1, w2, b)
plt.plot(p[0], p[1], marker='^', color = color, markersize=30)
```
